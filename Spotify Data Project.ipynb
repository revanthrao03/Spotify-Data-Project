```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
sns.countplot(x='genre', data = df)
```

```python
fig = px.histogram(df, x = 'genre', title = 'Distribution of Genre')
fig.update_xaxes(categoryorder = 'total ascending')
fig.update_xaxes(tickangle = -90)
fig.show()
```

```python
fig = px.box(df, x = 'year', y = 'danceability')
fig.show()
```

```python
df_year = df.groupby("year").mean().reset_index()
df_year

song_features = ["acousticness", "speechiness", "danceability", "energy", "instrumentalness", "liveness", "valence"]
fig = px.line(df_year, x = "year", y = song_features)
fig.show()
```

```python
fig = px.line(df_year, x = 'year', y = 'tempo')
fig.show()
```

```python
plt.figure(figsize=(12,8))

plt.scatter(data = df, x = 'valence', y = 'danceability')
plt.xlabel("Valence", fontsize = 15)
plt.ylabel("Danceability", fontsize = 15)
plt.title("Valence vs. Danceability", fontsize = 18)

x=df['valence']
y=df['danceability']
x.corr(y)
```

```python
fig = px.scatter(df, x='valence', y='danceability', title = 'Valence vs. Danceability', size = "acousticness")

fig.show()
```
